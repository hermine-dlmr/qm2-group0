```
#1 Timeline graph on F.I. 2019


import matplotlib.pyplot as plt
import pandas as pd

from google.colab import files
files.upload()
```

```
df = pd.read_csv('FI percentages data.csv')
plt.figure(figsize=(18,10))
plt.plot(df['Year'], df['Percentage FI individuals'], color='r', marker='o')
plt.grid(True)
plt.title('Evolution of the food insecurity rate in the US between 1998 and April 2020', fontsize=20)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Percentage of individuals who are food insecure', fontsize = 15)
```

```
#2 Timeline graph on F.I. 2020


import matplotlib.pyplot as plt
import pandas as pd

from google.colab import files
files.upload()
```

```
df1 = pd.read_csv('F.I. timeline data.csv')
df1.head()
```

```
plt.figure(figsize=(30,5))
plt.plot(df1['Week'], df1['Percentage of the population food insecure at a national level'], color='c', marker='o')
plt.grid(True)
plt.title('Evolution of the food insecurity rate in the US between the 23rd of April and the 7th of december 2020 in the US')
plt.xlabel('Week (starting from the 23rd of April)')
plt.ylabel('Percentage of individuals who are food insecure at a national level')
```

```
#interactive 
import plotly.express as px
timeline2 = px.line(df1, x = "Week", y = "Percentage of the population food insecure at a national level", title="Evolution of the food insecurity rate in the US between the 23rd of April and the 7th of december 2020 in the US",template="plotly")
timeline2.show()
```

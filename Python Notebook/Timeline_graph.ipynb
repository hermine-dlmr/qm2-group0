```
#1 Timeline graph on F.I. 2019


import matplotlib.pyplot as plt
import pandas as pd

from google.colab import files
files.upload()
```

```
df = pd.read_csv('FI percentages data (1).csv')
plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Percentage FI individuals'], color='c', marker='o')
plt.grid(True)
plt.title('Evolution of the food insecurity rate in the US between 1998 and 2019')
plt.xlabel('Year')
plt.ylabel('Percentage of individuals who are food insecure')
```

```
#2 Timeline graph on F.I. 2020


import matplotlib.pyplot as plt
import pandas as pd

from google.colab import files
files.upload()
```

```
df1 = pd.read_csv('F.I. .csv')
df1.head()
```

```
plt.figure(figsize=(30,5))
plt.plot(df1['Week'], df1['Percentage (B/C x 100)'], color='c', marker='o')
plt.grid(True)
plt.title('Evolution of the food insecurity rate in the US between the 23rd of April and the 7th of december 2020')
plt.xlabel('Week (starting from the 23rd of April')
plt.ylabel('Percentage of individuals who are food insecure')
```

```
#The code below will focus on creating Chloropleth maps 

#Two maps will be created on the food insecurity level per state - 1 with 2019 data - 2 with 2020 data (to draw comparisons)

#One map will be created on the racial demographics of US states with 2019 data
```

```
#1.1 2019 level of food insecurity 
import pandas as pd

df7 = pd.read_csv('/content/Book1.csv')
df7.head()
```

```
import pandas as pd
df7 = pd.read_csv('/content/Book1.csv')

import plotly.express as px 
fig = px.choropleth(df7, #the dataframe from which we will use the values
                    locations="Code", #the column with the USPS code
                    color="Percentage of the population who is food insecure in %", #the variable defining the range of colors
                    locationmode = 'USA-states') #a plot with all the us states

fig.update_layout(title_text = 'Level of food insecurity in each state in 2019', #the title of our plot 
                  geo_scope='usa', #we only want the USA and not the whole world
                  )
fig.show()  #to display the plot


```

```
#1.2 2020 level of food insecurity 
import pandas as pd

df8 = pd.read_csv('/content/data_2020.csv')
df8.head()
```

```
#we want to delete the unnamed column with no value inside
df8 = df8.loc[:, ~df8.columns.str.contains('^Unnamed')] #this piece of code is from 'https://www.codegrepper.com/code-examples/delphi/how+to+delete+Unnamed%3A+0+pandas'
df8.head()
```

```
import plotly.express as px 
fig = px.choropleth(df8, #the dataframe from which we will use the values
                    locations="USPS code", #the column with the USPS code
                    color='Percentage of the population food insecure in 2020', #the variable defining the range of colors
                    locationmode = 'USA-states') #a plot with all the us states

fig.update_layout(title_text = 'Level of food insecurity in each state in 2020', #the title of our plot 
                  geo_scope='usa', #we only want the USA and not the whole world
                  )
fig.show()  #to display the plot
```

```
#2 Racial demographic of US States 2019
import pandas as pd

df9 = pd.read_csv('/content/Book3.csv')
df9.head()
```

```
import plotly.express as px 
fig = px.choropleth(df9, #the dataframe from which we will use the values
                    locations="USPS code", #the column with the states 
                    color="All races except White", #the variable defining the range of colors
                    locationmode = 'USA-states') #a plot with all the us states

fig.update_layout(title_text = 'Racial ethnicities in the US in 2019', #the title of our plot 
                  geo_scope='usa', #we only want the USA and not the whole world
                  )
fig.show()  #to display the plot

```
